```python
car_sales=pd.read_csv('car-sales.csv')
car_sales.head()
```

```python
car_sales.info()
```

```python
plt.plot(car_sales['Price'])
plt.xlabel('Row')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(car_sales['Odometer (KM)'])
plt.xlabel('Row')
plt.ylabel('Odometer (KM)')
plt.show()
```

# You can't plot the price because it has data type not integer

```python
car_sales.columns
```

```python
car_sales["Price"] = car_sales["Price"].replace("[$,]", "", regex=True).astype(int)
car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.info()
```

```python
car_sales['Sales Date'] = pd.date_range(start ='2-11-2023', periods = len(car_sales))
car_sales.info()
```

```python
car_sales.head()
```

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs total sales line plot')
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer(KM) vs Price Scatter Plot')
```

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer(KM) Bar Plot')
```

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of car make')
plt.show()
```

```python
fig=plt.figure(figsize=(12,12))

ax1=fig.add_subplot(221)
ax1.set_title('Price vs. total sales line plot')

ax1.plot(car_sales['Price'],car_sales['Total Sales'])
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2=fig.add_subplot(222)
ax2.set_title('Odometer (KM) vs. Price scatter plot')

ax2.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')

ax3=fig.add_subplot(223)
ax3.set_title('Make vs. Odometer (KM) bar plot')

ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')

ax4=fig.add_subplot(224)
ax4.set_title('Histogram of car make')

ax4.hist(car_sales['Make'])
ax4.set_xlabel('Make')

plt.show()
```

```python
box_plot=sns.boxplot(x='Colour', 
                     y='Total Sales', 
                     data=car_sales)
```

```python
lm_plot=sns.lmplot(x='Total Sales', 
                   y='Odometer (KM)', 
                   data=car_sales, 
                   hue='Make', 
                   ci=False)
```

```python
pair_plot=sns.pairplot(car_sales)
```

```python
heat_map_corr=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
print(heat_map_corr)
cols=['Total Sales',
      'Price','Odometer (KM)']
heat_map=sns.heatmap(heat_map_corr,
                     annot=True,
                     xticklabels=cols,
                     yticklabels=cols)
```

## Between Price and Total Sales here is not really a big correlation. Between Odometer and Total Sales there is a slight negative correlation. Between Odometer and Price there is a medium negative correlation.
